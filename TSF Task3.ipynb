```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=data,palette='rocket',order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
data.hist(figsize=(10,10),bins=50)
plt.show()
```

```python
plt.figure(figsize=(15,12))
data['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
#Profit vs Discount
```

```python
from matplotlib import style
style.use('dark_background')
fig,ax=plt.subplots(figsize=(16,8))
ax.scatter(data['Sales'],data['Profit'])
ax.set_xlabel('Sales')
ax.set_xlabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['yellow','blue'],figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each region')
sns.barplot(x='Region',y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['red','skyblue'],figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```

```python
ps = data.groupby('State')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','blue'],figsize=(12,7))
plt.title('Profit/loss and Sales across states')
plt.xlabel('State')
plt.ylabel('Profit/loss and sales')
plt.show()
```

```python
top_states = data['State'].value_counts().nlargest(10)
top_states
```

```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','green'],alpha=0.9,figsize=(9,6))
plt.ylabel('Profit/loss and sales')
plt.show()
```

# Observation:
1.Technology and Office Supplies have high profit.

```python
ps = data.groupby('Sub-Category')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['yellow','blue'],figsize=(12,7))
plt.title('Profit/loss and Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss and sales')
plt.show()
```
